```python
# Import numpy and panda libraries to be able to do the necessary mathematical operations and import the nflfastR data

import numpy as np
import pandas as pd
YEAR = 2020
data = pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                   'play_by_play_' + str(YEAR) + '.csv.gz?raw=True', 
                   compression='gzip', low_memory=False)
```

```python
# Display the data to make sure everything looks good

data
```

```python
# Display each of the column headers in a list and identify which of the column headers I will be using for this program. Displaying all of them like this is much less time consuming than scrolling through the headers of the data

for col in data.columns:
    print(col)
```

```python
# Create a new dataset called thirdDown which will contain all the data from "data" but only include the 3rd down situations, which is all we need for this project

thirdDown = data.loc[(data.down == 3) & (data.third_down_converted != 'nan')]
```

```python
# Display this new dataset and make sure everything looks good

thirdDown
```

```python
# To see the data type for the cells in those particular columns and identification for converting/failing using binary number system which I will use throughout the rest of the code

print(data['third_down_failed'].unique())
print(data['third_down_converted'].unique())
print(data['ydstogo'].unique())
print(data['play_type'].unique())
```

```python
# How many total shotgun plays were run in 2020 season

shotgun = thirdDown.loc[(thirdDown.shotgun == 1)]
print('Shotgun: ' + str(len(shotgun)))
```

```python
# Total 3rd Down Conversion Percentage

thirdDownConvert = thirdDown.loc[(thirdDown.third_down_converted == 1)]
thirdDownFail = thirdDown.loc[(thirdDown.third_down_failed == 1)]
conversionPercent = len(thirdDownConvert) / (len(thirdDownConvert) + len(thirdDownFail))*100
print('3rd down Total Conversion Percentage: ' + str(conversionPercent) + ' %')
```

```python
# Number of 3rd downs in each yards-to-go situation 2020

veryShort = thirdDown.loc[(thirdDown.ydstogo <= 1)]
short = thirdDown.loc[(thirdDown.ydstogo >= 2) & (thirdDown.ydstogo <= 3)]
medium = thirdDown.loc[(thirdDown.ydstogo >= 4) & (thirdDown.ydstogo <= 7)]
long = thirdDown.loc[(thirdDown.ydstogo >= 8)]
print('Number of Very Short 3rd down ytg (<= 1 yd): ' + str(len(veryShort)))
print('Number of Short 3rd down ytg (2-3 yd): ' + str(len(short)))
print('Number of Medium 3rd down ytg (4-7 yd): ' + str(len(medium)))
print('Number of Long 3rd down ytg (8+ yd): ' + str(len(long)))
```

```python
# Aggregated average of an NFL team running shotgun formation in each 3rd down and yards-to-go situation in 2020 season

veryShortShotgun = veryShort.loc[(veryShort.shotgun == 1)]
print('2020 NFL Percentage of running Shotgun Formation on 3rd and Very Short: ' + str(len(veryShortShotgun)/len(veryShort)*100) + ' %')
shortShotgun = short.loc[(short.shotgun == 1)]
print('2020 NFL Percentage of running Shotgun Formation on 3rd and Short: ' + str(len(shortShotgun)/len(short)*100) + ' %') 
mediumShotgun = medium.loc[(medium.shotgun == 1)]
print('2020 NFL Percentage of running Shotgun Formation on 3rd and Very Short: ' + str(len(mediumShotgun)/len(medium)*100) + ' %')
longShotgun = long.loc[(long.shotgun == 1)]
print('2020 NFL Percentage of running Shotgun Formation on 3rd and Very Short: ' + str(len(longShotgun)/len(long)*100) + ' %')
```

```python
# Conversion percentage in each 3rd down and yards-to-go situation in 2020 season

convertVeryShortShotgun = veryShortShotgun.loc[(veryShortShotgun.third_down_converted == 1)]
print('Conversion Percentage of Shotgun Formation on 3rd and Very Short: ' + str(len(convertVeryShortShotgun)/len(veryShortShotgun)*100) + ' %')
veryShortNonShotgun = veryShort.loc[(veryShort.shotgun == 0)]
convertVeryShortNonShotgun = veryShortNonShotgun.loc[(veryShortNonShotgun.third_down_converted == 1)]
print('Conversion Percentage of Non-Shotgun Formation on 3rd and Very Short: ' + str(len(convertVeryShortNonShotgun)/len(veryShortNonShotgun)*100) + ' %')
convertShortShotgun = shortShotgun.loc[(shortShotgun.third_down_converted == 1)]
print('Conversion Percentage of Shotgun Formation on 3rd and Short: ' + str(len(convertShortShotgun)/len(shortShotgun)*100) + ' %')
shortNonShotgun = short.loc[(short.shotgun == 0)]
convertShortNonShotgun = shortNonShotgun.loc[(shortNonShotgun.third_down_converted == 1)]
print('Conversion Percentage of Non-Shotgun Formation on 3rd and Short: ' + str(len(convertShortNonShotgun)/len(shortNonShotgun)*100) + ' %')
convertMediumShotgun = mediumShotgun.loc[(mediumShotgun.third_down_converted == 1)]
print('Conversion Percentage of Shotgun Formation on 3rd and Medium: ' + str(len(convertMediumShotgun)/len(mediumShotgun)*100) + ' %')
mediumNonShotgun = medium.loc[(medium.shotgun == 0)]
convertMediumNonShotgun = mediumNonShotgun.loc[(mediumNonShotgun.third_down_converted == 1)]
print('Conversion Percentage of Non-Shotgun Formation on 3rd and Medium: ' + str(len(convertMediumNonShotgun)/len(mediumNonShotgun)*100) + ' %')
convertLongShotgun = longShotgun.loc[(longShotgun.third_down_converted == 1)]
print('Conversion Percentage of Shotgun Formation on 3rd and Medium: ' + str(len(convertLongShotgun)/len(longShotgun)*100) + ' %')
longNonShotgun = long.loc[(long.shotgun == 0)]
convertLongNonShotgun = longNonShotgun.loc[(longNonShotgun.third_down_converted == 1)]
print('Conversion Percentage of Non-Shotgun Formation on 3rd and Medium: ' + str(len(convertLongNonShotgun)/len(longNonShotgun)*100) + ' %')
```

```python
# Conversion percentage of passing the ball and running the ball in each 3rd down and yards-to-go situation in 2020 season

passConvertVeryShortShotgun = convertVeryShortShotgun.loc[(convertVeryShortShotgun.play_type == 'pass')]
passVeryShortShotgun = veryShortShotgun.loc[(veryShortShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Shotgun Formation on 3rd and Very Short: ' + str(len(passConvertVeryShortShotgun)/len(passVeryShortShotgun)*100) + ' %')
runConvertVeryShortShotgun = convertVeryShortShotgun.loc[(convertVeryShortShotgun.play_type == 'run')]
runVeryShortShotgun = veryShortShotgun.loc[(veryShortShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Very Short: ' + str(len(runConvertVeryShortShotgun)/len(runVeryShortShotgun)*100) + ' %')
passConvertVeryShortNonShotgun = convertVeryShortNonShotgun.loc[(convertVeryShortNonShotgun.play_type == 'pass')]
passVeryShortNonShotgun = veryShortNonShotgun.loc[(veryShortNonShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Non-Shotgun Formation on 3rd and Very Short: ' + str(len(passConvertVeryShortNonShotgun)/len(passVeryShortNonShotgun)*100) + ' %')
runConvertVeryShortNonShotgun = convertVeryShortNonShotgun.loc[(convertVeryShortNonShotgun.play_type == 'run')]
runVeryShortNonShotgun = veryShortNonShotgun.loc[(veryShortNonShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Non-Shotgun Formation on 3rd and Very Short: ' + str(len(runConvertVeryShortNonShotgun)/len(runVeryShortNonShotgun)*100) + ' %')
passConvertShortShotgun = convertShortShotgun.loc[(convertShortShotgun.play_type == 'pass')]
passShortShotgun = shortShotgun.loc[(shortShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Shotgun Formation on 3rd and Short: ' + str(len(passConvertShortShotgun)/len(passShortShotgun)*100) + ' %')
runConvertShortShotgun = convertShortShotgun.loc[(convertShortShotgun.play_type == 'run')]
runShortShotgun = shortShotgun.loc[(shortShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Short: ' + str(len(runConvertShortShotgun)/len(runShortShotgun)*100) + ' %')
passConvertShortNonShotgun = convertShortNonShotgun.loc[(convertShortNonShotgun.play_type == 'pass')]
passShortNonShotgun = shortNonShotgun.loc[(shortNonShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Non-Shotgun Formation on 3rd and Short: ' + str(len(passConvertShortNonShotgun)/len(passShortNonShotgun)*100) + ' %')
runConvertShortNonShotgun = convertShortNonShotgun.loc[(convertShortNonShotgun.play_type == 'run')]
runShortNonShotgun = shortNonShotgun.loc[(shortNonShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Non-Shotgun Formation on 3rd and Short: ' + str(len(runConvertShortNonShotgun)/len(runShortNonShotgun)*100) + ' %')
passConvertMediumShotgun = convertMediumShotgun.loc[(convertMediumShotgun.play_type == 'pass')]
passMediumShotgun = mediumShotgun.loc[(mediumShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Shotgun Formation on 3rd and Medium: ' + str(len(passConvertMediumShotgun)/len(passMediumShotgun)*100) + ' %')
runConvertMediumShotgun = convertMediumShotgun.loc[(convertMediumShotgun.play_type == 'run')]
runMediumShotgun = mediumShotgun.loc[(mediumShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Medium: ' + str(len(runConvertMediumShotgun)/len(runMediumShotgun)*100) + ' %')
passConvertMediumNonShotgun = convertMediumNonShotgun.loc[(convertMediumNonShotgun.play_type == 'pass')]
passMediumNonShotgun = mediumNonShotgun.loc[(mediumNonShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Non-Shotgun Formation on 3rd and Medium: ' + str(len(passConvertMediumNonShotgun)/len(passMediumNonShotgun)*100) + ' %')
runConvertMediumNonShotgun = convertMediumNonShotgun.loc[(convertMediumNonShotgun.play_type == 'run')]
runMediumNonShotgun = mediumNonShotgun.loc[(mediumNonShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Non-Shotgun Formation on 3rd and Medium: ' + str(len(runConvertMediumNonShotgun)/len(runMediumNonShotgun)*100) + ' %')
passConvertLongShotgun = convertLongShotgun.loc[(convertLongShotgun.play_type == 'pass')]
passLongShotgun = longShotgun.loc[(longShotgun.play_type == 'pass')]
print('Conversion Percentage of Passing in Shotgun Formation on 3rd and Long: ' + str(len(passConvertLongShotgun)/len(passLongShotgun)*100) + ' %')
runConvertLongShotgun = convertLongShotgun.loc[(convertLongShotgun.play_type == 'run')]
runLongShotgun = longShotgun.loc[(longShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Long: ' + str(len(runConvertLongShotgun)/len(runLongShotgun)*100) + ' %')
passConvertLongNonShotgun = convertLongNonShotgun.loc[(convertLongNonShotgun.play_type == 'pass')]
passLongNonShotgun = longNonShotgun.loc[(longNonShotgun.play_type == 'pass')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Long: ' + str(len(passConvertLongNonShotgun)/len(passLongNonShotgun)*100) + ' %')
runConvertLongNonShotgun = convertLongNonShotgun.loc[(convertLongNonShotgun.play_type == 'run')]
runLongNonShotgun = longNonShotgun.loc[(longNonShotgun.play_type == 'run')]
print('Conversion Percentage of Running in Shotgun Formation on 3rd and Long: ' + str(len(runConvertLongNonShotgun)/len(runLongNonShotgun)*100) + ' %')
```
